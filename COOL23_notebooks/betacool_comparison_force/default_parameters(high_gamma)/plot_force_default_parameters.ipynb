```python
import numpy as np
import matplotlib.pyplot as plt



def load_data(module):
    filename = f"results/force_{module}.npz"
    data = np.load(filename)
    v_diff = data['v_diff']
    force = data['force']
    return v_diff,force

cmap = plt.get_cmap("tab10")
plt.figure(figsize=(12, 8))
plt.rcParams.update({'font.size': 16})


v_diff,force = load_data('betacool')
plt.plot(v_diff, -force, label=f'betacool',color=cmap(0),linestyle='--')
v_diff,force= load_data('xsuite')
plt.plot(v_diff, -force, label=f'xsuite',color=cmap(0))

plt.xlabel('$\Delta$v [m/s]')
plt.ylabel('-Force [eV/m]')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.legend()
plt.xlim([0,3e5])
plt.axis(ymin = 0)
plt.savefig('force_default_parameters.png', dpi=300)
plt.savefig('force_default_parameters.eps', dpi=300)
plt.show()
```
