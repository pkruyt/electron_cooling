```python
# copyright ############################### #
# This file is part of the Xtrack Package.  #
# Copyright (c) CERN, 2021.                 #
# ######################################### #


import xtrack as xt
import xobjects as xo
import xpart as xp

import ducktrack as dtk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from tqdm import tqdm

# Load and unpack the first file
data_betacool_A = np.load('results/emittance_betacool_L=1.npz')
emittance_betacool_A = data_betacool_A['emittance_x']
time_betacool_A = data_betacool_A['time']


# # Load and unpack the second file
data_xsuite_A = np.load('results/emittance_xsuite_L=1.npz')
emittance_xsuite_A = data_xsuite_A['emittance']
time_xsuite_A = data_xsuite_A['time']

# Load and unpack the third file
data_betacool_B = np.load('results/emittance_betacool_L=10.npz')
emittance_betacool_B = data_betacool_B['emittance_x']
time_betacool_B = data_betacool_B['time']


# Load and unpack the fourth file
data_xsuite_B = np.load('results/emittance_xsuite_L=10.npz')
emittance_xsuite_B = data_xsuite_B['emittance']
time_xsuite_B = data_xsuite_B['time']
```

```python
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(8, 6))

plt.plot(time_xsuite_A, emittance_xsuite_A, color='blue',       label='Xsuite    :L=1 m')
plt.plot(time_betacool_A, emittance_betacool_A, color='red', label='Betacool:L=1 m')

plt.plot(time_xsuite_B, emittance_xsuite_B, color='purple', label='Xsuite    :L=10 m')
plt.plot(time_betacool_B, emittance_betacool_B, color='orange', label='Betacool:L=10 m')


plt.legend()
plt.ylabel(r'$\epsilon$ [$\mu$ rad]')
plt.xlabel('Time [s]')

# plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
# plt.ticklabel_format(useOffset=False,style='plain')
plt.subplots_adjust(bottom=0.15)
plt.savefig('emittance-length.png', dpi=300)
plt.savefig('emittance-length.eps', dpi=300)

plt.show()
```
