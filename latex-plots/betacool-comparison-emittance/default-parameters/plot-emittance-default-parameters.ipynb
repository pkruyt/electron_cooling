```python
# copyright ############################### #
# This file is part of the Xtrack Package.  #
# Copyright (c) CERN, 2021.                 #
# ######################################### #


import xtrack as xt
import xobjects as xo
import xpart as xp

import ducktrack as dtk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
from tqdm import tqdm

filepath_xsuite='results/emittance_xsuite.npz'

loaded_data = np.load(filepath_xsuite)
time_xs = loaded_data['time']
emittance_xs = loaded_data['emittance']*1e6

filepath_betacool='results/emittance_betacool.npz'

loaded_data = np.load(filepath_betacool)
time_betacool = loaded_data['time']
emittance_x_betacool = loaded_data['emittance_x']*1e6
# emittance_y_betacool = loaded_data['emittance_y']
```

```python
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(8, 6))

plt.plot(time_betacool,emittance_x_betacool,label='Betacool')
plt.plot(time_xs,emittance_xs,label='Xsuite')


plt.legend()
plt.ylabel(r'$\epsilon$ [$\mu$ rad]')
plt.xlabel('Time [s]')

# plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
# plt.ticklabel_format(useOffset=False,style='plain')
plt.subplots_adjust(bottom=0.15)
plt.savefig('emittance-default-parameters.png', dpi=300)
plt.savefig('emittance-default-parameters.eps', dpi=300)

plt.show()
```
