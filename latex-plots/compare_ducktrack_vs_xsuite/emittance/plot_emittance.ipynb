```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data from the .npz file
filepath = 'results/emittance_xsuite.npz'
data_xsuite = np.load(filepath)

time = data_xsuite['time']
emittance_x_xsuite = data_xsuite['emittance_x']
emittance_y_xsuite = data_xsuite['emittance_y']
dp_p_list = data_xsuite['dp_p']

# Load the data from the .npz file
filepath = 'results/emittance_ducktrack.npz'
data_ducktrack= np.load(filepath)

time = data_ducktrack['time']
emittance_x_ducktrack = data_ducktrack['emittance_x']
emittance_y_ducktrack = data_ducktrack['emittance_y']
dp_p_list = data_ducktrack['dp_p']

# Plot the results
plt.figure(figsize=(10, 6))

# Plot emittance_x
plt.plot(time, emittance_x_xsuite, label='xsuite')
plt.plot(time, emittance_x_ducktrack, label='ducktrack')
plt.xlabel('Time')
plt.ylabel('Emittance_x')
plt.legend()
plt.tight_layout()
plt.show()
```
