```python
import numpy as np
import matplotlib.pyplot as plt

# Load and unpack the first file
data_betacool_1 = np.load('results/friction_force_betacool_Q=40.npz')
force_betacool_1 = data_betacool_1['force']


# Load and unpack the second file
data_xsuite_1 = np.load('results/friction-force-xsuite_Q=40.npz')
v_diff_xsuite_1 = data_xsuite_1['v_diff']
force_xsuite_1 = data_xsuite_1['force']


# Load and unpack the third file
data_betacool_2 = np.load('results/friction_force_betacool_Q=54.npz')
force_betacool_2 = data_betacool_2['force']


# Load and unpack the fourth file
data_xsuite_2 = np.load('results/friction-force-xsuite_Q=54.npz')
v_diff_xsuite_2 = data_xsuite_2['v_diff']
force_xsuite_2 = data_xsuite_2['force']


```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(8, 6))

plt.scatter(v_diff_xsuite_1, -force_xsuite_1, color='blue',       label='Xsuite    :Q=40 e')
plt.plot(force_betacool_1[:, 0], -force_betacool_1[:, 1], color='red', label='Betacool:Q=40 e')

plt.scatter(v_diff_xsuite_2, -force_xsuite_2, color='purple', label='Xsuite    :Q=54 e')
plt.plot(force_betacool_2[:, 0], -force_betacool_2[:, 1], color='orange', label='Betacool:Q=54 e')

plt.xlim([-6 * 1e5, 6 * 1e5])
plt.legend()
plt.ylabel('-F Longitudinal [eV/m]')
plt.xlabel('Δv [m/s]')
plt.ticklabel_format(axis="x", style="sci", scilimits=(0, 0))

plt.ticklabel_format(useOffset=False, style='plain')
plt.subplots_adjust(bottom=0.15)
plt.savefig('force-charge.png', dpi=300)

plt.show()

```
