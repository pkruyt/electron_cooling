```python
import numpy as np
import matplotlib.pyplot as plt

# Load and unpack the first file
data_betacool_0_1 = np.load('results/friction_force_betacool_B=0.1.npz')
force_betacool_0_1 = data_betacool_0_1['force']


# Load and unpack the second file
data_xsuite_0_1 = np.load('results/friction-force-xsuite_B=0.1.npz')
v_diff_xsuite_0_1 = data_xsuite_0_1['v_diff']
force_xsuite_0_1 = data_xsuite_0_1['force']


# Load and unpack the third file
data_betacool_0_001 = np.load('results/friction_force_betacool_B=0.001.npz')
force_betacool_0_001 = data_betacool_0_001['force']


# Load and unpack the fourth file
data_xsuite_0_001 = np.load('results/friction-force-xsuite_B=0.001.npz')
v_diff_xsuite_0_001 = data_xsuite_0_001['v_diff']
force_xsuite_0_001 = data_xsuite_0_001['force']


```

```python
plt.rcParams.update({'font.size': 14})
plt.figure(figsize=(8, 6))

plt.scatter(v_diff_xsuite_0_1, -force_xsuite_0_1, color='blue',       label='Xsuite    :B=0.1 T')
plt.plot(force_betacool_0_1[:, 0], -force_betacool_0_1[:, 1], color='red', label='Betacool:B=0.1 T')

plt.scatter(v_diff_xsuite_0_001, -force_xsuite_0_001, color='purple', label='Xsuite    :B=0.001 T')
plt.plot(force_betacool_0_001[:, 0], -force_betacool_0_001[:, 1], color='orange', label='Betacool:B=0.001 T')

plt.xlim([-6 * 1e5, 6 * 1e5])
plt.legend()
plt.ylabel('-F Longitudinal [eV/m]')
plt.xlabel('v [m/s]')
plt.ticklabel_format(axis="x", style="sci", scilimits=(0, 0))

plt.ticklabel_format(useOffset=False, style='plain')
plt.subplots_adjust(bottom=0.15)
plt.savefig('force-magnetic-field.png', dpi=300)

plt.show()

```
