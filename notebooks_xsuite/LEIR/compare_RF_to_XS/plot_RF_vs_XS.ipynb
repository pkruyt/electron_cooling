```python
import numpy as np
import matplotlib.pyplot as plt

sec_per_turn=1.9951813726709976e-06

# Specify the file paths
file_path_no_cooler_xs_rf = 'results/rf_no_cooler.npz'
file_path_cooler_rf = 'results/rf_cooler.npz'

# Load the results from the files
results_no_cooler_rf = np.load(file_path_no_cooler_xs_rf)
results_cooler_rf = np.load(file_path_cooler_rf)

# Access the loaded data
x_rf_no_cooler = results_no_cooler_rf['x_rf']
px_rf_no_cooler = results_no_cooler_rf['px_rf']
hist_1_rf_no_cooler = results_no_cooler_rf['hist_1_rf']
hist_2_rf_no_cooler = results_no_cooler_rf['hist_2_rf']
width_rf_no_cooler = results_no_cooler_rf['width_rf']

x_rf_cooler = results_cooler_rf['x_rf']
px_rf_cooler = results_cooler_rf['px_rf']
hist_1_rf_cooler = results_cooler_rf['hist_1_rf']
hist_2_rf_cooler = results_cooler_rf['hist_2_rf']
width_rf_cooler = results_cooler_rf['width_rf']

print(x_rf_cooler.shape[0])
print(x_rf_no_cooler.shape[0])
```

```python
file_path_no_cooler_xs = 'results/xsuite_no_cooler.npz'
file_path_cooler_xs = 'results/xsuite_cooler.npz'

# Load the results from the files
results_no_cooler_xs = np.load(file_path_no_cooler_xs)
results_cooler_xs = np.load(file_path_cooler_xs)

# Access the loaded data
x_xs_no_cooler = results_no_cooler_xs['x_xs']
px_xs_no_cooler = results_no_cooler_xs['px_xs']
hist_1_xs_no_cooler = results_no_cooler_xs['hist_1_xs']
hist_2_xs_no_cooler = results_no_cooler_xs['hist_2_xs']
width_xs_no_cooler = results_no_cooler_xs['width_xs']

x_xs_cooler = results_cooler_xs['x_xs']
px_xs_cooler = results_cooler_xs['px_xs']
hist_1_xs_cooler = results_cooler_xs['hist_1_xs']
hist_2_xs_cooler = results_cooler_xs['hist_2_xs']
width_xs_cooler = results_cooler_xs['width_xs']

interval=100
time = np.arange(0, len(width_xs_no_cooler)*interval, interval) * sec_per_turn

time_cool = np.arange(0, len(width_xs_cooler)*interval, interval) * sec_per_turn
```

```python
plt.figure()
plt.title('Xsuite vs RF Track')
plt.plot(time,hist_1_xs_no_cooler,label='Xsuite without cooler')
plt.plot(time,hist_2_xs_no_cooler,label='Xsuite without cooler')
plt.plot(time,hist_1_rf_no_cooler,label='RF Track without cooler')
plt.plot(time,hist_2_rf_no_cooler,label='Xsuite without cooler')
plt.xlabel('time [s]')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.title('Xsuite vs RF Track')
plt.plot(time,width_rf_no_cooler,label='RF Track without cooler')
plt.plot(time,width_xs_no_cooler,label='Xsuite without cooler')
plt.xlabel('time [s]')
plt.legend()
plt.show()
```

```python


plt.figure()
plt.title('Xsuite')
plt.plot(time_cool,width_xs_cooler,label='Xsuite with cooler')
#plt.plot(time,width_xs_no_cooler,label='Xsuite without cooler')
plt.xlabel('time [s]')
plt.legend()
plt.show()

plt.figure()
plt.title('RF Track')
plt.plot(time_cool,width_rf_cooler,label='Width')
plt.plot(time_cool,hist_1_rf_cooler,label='Hist 1')
plt.plot(time_cool,hist_2_rf_cooler,label='Hist 2')

#plt.plot(width_rf_no_cooler,label='RF Track without cooler')
plt.xlabel('time [s]')
plt.legend()
plt.show()
```
