```python
import matplotlib.pyplot as plt
import numpy as np

# Load the variables from the first .npz file
data_xs = np.load('emittance_xs.npz')
time_xs = data_xs['time']
emittance_xs_list = data_xs['emittance']*1e6

# Load the variables from the second .npz file
data_rf = np.load('emittance_rf.npz')
time_rf = data_rf['time']
emitt_4d_list = data_rf['emittance']*1e6


```

```python
plt.rcParams.update({'font.size': 16})
plt.figure(figsize=(8, 6))

plt.plot(time_xs,emittance_xs_list,label='Xsuite    :  emittance 2d')
plt.plot(time_rf,emitt_4d_list,label='RF Track : emittance 4d')

plt.legend()
plt.ylabel(r'$\epsilon$ [$\mu$ rad]')
plt.xlabel('Time [s]')

# plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))
# plt.ticklabel_format(useOffset=False,style='plain')
plt.subplots_adjust(bottom=0.15)
plt.savefig('emittance-rf-track.png', dpi=300)
plt.savefig('emittance-rf-track.eps', dpi=300)

plt.show()
```
