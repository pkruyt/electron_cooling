```python
import numpy as np
import matplotlib.pyplot as plt

file = 'action_reduction_data_pos.npz'
data = np.load(file)

disp_values = data['disp_values']
offset_values = data['offset_values']
action_reduction = data['action_reduction']

plt.figure(figsize=(10, 6))
im = plt.imshow(action_reduction, cmap='hot', aspect='auto', origin='lower',
                extent=[offset_values[0], offset_values[-1], disp_values[0], disp_values[-1]])

vmin = np.min(action_reduction)


# Set the upper limit of the colorbar to 1
im.set_clim(vmin=vmin, vmax=1)

# Create a custom colormap that sets values above 1 to white
cmap = plt.cm.get_cmap('hot')
cmap.set_over('white')

# Create the colorbar with the modified colormap
cbar = plt.colorbar(im, label='Action reduction (Temperature)')

# Set the colorbar ticks and tick labels
cbar.set_ticks([vmin, (1+vmin)/2, 1])
cbar.set_ticklabels([f'{vmin:.2f}', f'{(1+vmin)/2:.2f}', '1'])

plt.ylabel('Disp Values')
plt.xlabel('Energy offset [eV]')
plt.title('Action reduction heatmap (positive dp/p)')
plt.show()
```
