```python
import numpy as np
import matplotlib.pyplot as plt


file='cooling_time_data_pos.npz'

data = np.load(file)

import matplotlib.pyplot as plt
import numpy as np

# Assuming you have already loaded the data into the following arrays
disp_values = data['disp_values']
offset_values = data['offset_values']
cooling_time = data['cooling_time']

# Reshape cooling_time list to a 20x10 array
# cooling_time = np.reshape(cooling_time, (10, 10))

# Create a heatmap
plt.figure(figsize=(10, 6))
plt.imshow(cooling_time, cmap='hot', aspect='auto', origin='lower',
           extent=[offset_values[0], offset_values[-1], disp_values[0], disp_values[-1]])
plt.colorbar(label='Cooling Time (Temperature)')
plt.xlabel('Disp Values')
plt.ylabel('Offset_dp Values')
plt.title('Cooling Time Heatmap')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

file1 = 'cooling_time_data_pos.npz'
file2 = 'cooling_time_data_neg.npz'

data1 = np.load(file1)
data2 = np.load(file2)

cooling_time1 = data1['cooling_time']
cooling_time2 = data2['cooling_time']

cooling_time_tot = cooling_time1 + cooling_time2

offset_values = np.linspace(-1e-9, 1e-9, num=30)  # Adjusted range to include zero
offset_slice=5e-9

offset_index = np.abs(offset_values - offset_slice).argmin()  # Find the index for the nearest value to zero

# Select only the values for offset=0
cooling_time_offset_0 = cooling_time_tot[:, offset_index]

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(disp_values, cooling_time_offset_0)
plt.xlabel('Disp Values')
plt.ylabel('Cooling Time (Temperature)')
plt.title(f'Cooling Time for Offset = {offset_slice}')
plt.show()


```
